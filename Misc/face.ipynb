```python
from PIL import Image
import numpy as np
import matplotlib.pyplot as plt
import cv2 as cv
```

```python
frame = cv.imread("bruh.jpg")
colour = cv.cvtColor(frame, cv.COLOR_BGR2RGB)
```

```python
plt.imshow(colour)
```

```python
classifier = cv.CascadeClassifier("haarcascade_frontalface_default.xml")
```

```python
faces = classifier.detectMultiScale(frame)
```

```python
face = faces[1]
```

```python
x,y,w,h = face
```

```python
face = colour[y:y+h, x:x+w]
```

```python
plt.imshow(face)
```

```python
out = cv2.rectangle(colour, (x,y), (x+w,y+h), (0,0,255), 4)
```

```python
plt.imshow(out)
```
